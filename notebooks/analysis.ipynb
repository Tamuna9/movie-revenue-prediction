```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("../data/tmdb_5000_movies.csv")

df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
plt.figure(figsize=(8,5))
sns.histplot(df["revenue"], kde=True)
plt.title("Revenue Distribution")
plt.xlabel("Revenue")
plt.ylabel("Count")
plt.tight_layout()
plt.savefig("../images/revenue_distribution.png")
plt.show()
```

```python
sns.scatterplot(data=df, x="budget", y="revenue")
plt.savefig("images/rating_vs_revenue.png")
plt.show()
```

```python
numeric = df[[
    "budget",
    "revenue",
    "runtime",
    "vote_average",
    "vote_count",
    "popularity"
]]

sns.heatmap(numeric.corr(), annot=True, cmap="coolwarm")
plt.savefig("..\images\correlation_heatmap.png")
plt.show()
```

```python
sns.histplot(df["budget"], bins=40, kde=True)

plt.title("Distribution of Movie Budgets")
plt.xlabel("Budget")
plt.ylabel("Count")
plt.savefig("../images/revenue_distribution.png")
plt.show()
```

```python
sns.scatterplot(data=df, x="vote_average", y="revenue")

plt.title("Rating vs Revenue")
plt.savefig("../images/rating_vs_revenue.png")
plt.show()
```

```python
plt.figure(figsize=(10,6))

sns.regplot(
    data=df,
    x="budget",
    y="revenue",
    scatter_kws={"alpha":0.3}
)

plt.title("Budget vs Revenue Trend")
plt.savefig("../images/budget_vs_revenue.png")
plt.show()
```
